## Annotations statistics

In this section we reproduce Table 1 of our paper which displays stats concerning the
annotations performed in the different proposals.

Running this code block may take time (~10 minutes).

```python
stats = annotation_stats.compute_stats()

table = """
| **Metric** | **Value** |
|:-----------|----------:|"""

for k, v in stats.items():
    table += f"\n| {k} | {v} |"

display(Markdown(table))
```
